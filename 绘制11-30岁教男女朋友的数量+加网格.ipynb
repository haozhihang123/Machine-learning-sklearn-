```python
from matplotlib import pyplot as plt
import random
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

```python
y = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]
x = range(11,31)
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
plt.xticks(list(x)[::1],rotation=30) 
#添加描述信息
plt.xlabel("年龄")
plt.ylabel("次数")
plt.title("11-30岁交男女朋友的次数统计图")
plt.grid(alpha=0.4) # alpha设置网格透明度
plt.show()
```
