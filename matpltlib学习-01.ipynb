```python
from matplotlib import pyplot as plt
```

```python
# 设置图片大小
plt.figure(figsize=(20,8),dpi=80)
```

```python
# 创建数据
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,27,22,18,15]
```

```python
# 绘图
plt.plot(x,y)
# 绘制x轴的刻度
# plt.xticks(range(2,25))
_x_lable = [i/2 for i in range(4,49)]
print(_x_lable[::4])
plt.xticks(_x_lable[::4])
plt.yticks(range(min(y),max(y)))
# # 图片保存
# plt.savefig("./t1.png")
```

```python
# 显示所绘图形
plt.show()
```
